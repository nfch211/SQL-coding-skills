```python
from itertools import combinations
import pandas as pd


#changed from alphabets to numbers so that numbers could now be used as a sign to track the right position by counting whether the difference in between the back and front numbers is 0.
#numbers are going to be assigned to any characters in any address.

str1 ="123456789011"
str2 ="645656789065"


#for str1, 
combine1 = {}
for i in range(1, len(str1)):   
    combine1.update({i:list(combinations(str1, i+1))})
    
combine2 = {}
for j in range(1, len(str2)):
    combine2.update({j:list(combinations(str2, j+1))})


collections1 = [value for sublist in combine1.values() for value in sublist]

collections2 = [value for sublist in combine2.values() for value in sublist]

common_matching = []
for k in collections1:
    if k in collections2:
        common_matching.append(k)


str_dic = {index:w for index, w in enumerate(str1)}
    
for it in common_matching:
    for key, value in str_dic.items():
        if value in it:
            print(value, key, it) 
            

combine1 = {}      
```
